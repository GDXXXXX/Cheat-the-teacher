```python
# 
# 
import pandas as pd
import matplotlib.pyplot as plt

# 
ts_data = pd.read_csv('ST(000616).csv',encoding='gbk')

# DataFrame
df1 = pd.DataFrame(ts_data)

df1.duplicated().sum()
df1.isna().sum()

# 
df1 = df1.drop_duplicates()
# 
df1 = df1.dropna(axis=0, how='all')

# 10
df1 = df1.iloc[:, 0:10]

# 
df1.index = df1['']
df1 = df1.drop([''], axis=1)

#X,Y
x=[x for x in df1.columns][::-1]
y=[float(y) for y in df1.loc['(%)']][::-1]
y1=[float(y) for y in df1.loc['(%)']][::-1]
y2=[float(y) for y in df1.loc['(%)']][::-1]
y3=[float(y) for y in df1.loc['(%)']][::-1]
y4=[float(y) for y in df1.loc['(%)']][::-1]
y5=[float(y) for y in df1.loc['(%)']][::-1]

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
plt.subplots(1, 1, figsize=(10, 6), dpi=200)

# 
plt.plot(x, y, color='blue', marker='o', linestyle='-', alpha=0.5)
plt.plot(x, y1, color='red', marker='o', linestyle='-', alpha=0.5)
plt.plot(x, y2, color='y', marker='o', linestyle='-', alpha=0.5)
plt.plot(x, y3, color='black', marker='o', linestyle='-', alpha=0.5)
plt.plot(x, y4, color='purple', marker='o', linestyle='-', alpha=0.5)
plt.plot(x, y5, color='g', marker='o', linestyle='-', alpha=0.5)
plt.legend(['', '','','','',''])  # 
# 
plt.xlabel('', fontsize='16')
plt.ylabel('', fontsize='16')
# plt.yticks(range(0,50,5))

# 
for x_, y_ in zip(x, y):
    plt.text(x_, y_, y_, ha='left', va='bottom')
for x_, y_ in zip(x, y1):
    plt.text(x_, y_, y_, ha='left', va='bottom')
for x_, y_ in zip(x, y2):
    plt.text(x_, y_, y_, ha='left', va='bottom')
for x_, y_ in zip(x, y3):
    plt.text(x_, y_, y_, ha='left', va='bottom')
for x_, y_ in zip(x, y4):
    plt.text(x_, y_, y_, ha='left', va='bottom')
for x_, y_ in zip(x, y5):
    plt.text(x_, y_, y_, ha='left', va='bottom')

```
