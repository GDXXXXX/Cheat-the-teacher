```python
# 导入库
import pandas as pd
import matplotlib.pyplot as plt

# 获取数据
ts_data = pd.read_csv('上海市历年居民消费价格指数.csv', encoding='gbk')

# 将数据转换为DataFrame数据
df1 = pd.DataFrame(ts_data)

df1.duplicated().sum()
df1.isna().sum()

# 重复数据处理
df1 = df1.drop_duplicates()
# 空值处理
df1 = df1.dropna(axis=1, how='all')
# # 获取最近10期报告数据
df1 = df1[:15]

#设置X轴,Y轴数据
x=[x for x in df1['时间']][::-1]
y=[float(y) for y in df1['指标']][::-1]

# 解决汉字显示乱码
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置画布大小
plt.subplots(1, 1, figsize=(10, 6), dpi=100)

# 折线图
plt.plot(x, y, color='red', marker='o', linestyle='-', alpha=0.5)

# 添加标签
plt.xlabel('时间', fontsize='16')
plt.ylabel('居民消费价格指数', fontsize='16')

# 对每个数据点加标注
for x_, y_ in zip(x, y):
    plt.text(x_, y_, y_, ha='left', va='bottom')
```
