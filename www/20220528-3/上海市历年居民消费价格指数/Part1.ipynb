```python
# 
import pandas as pd
import matplotlib.pyplot as plt

# 
ts_data = pd.read_csv('.csv', encoding='gbk')

# DataFrame
df1 = pd.DataFrame(ts_data)

df1.duplicated().sum()
df1.isna().sum()

# 
df1 = df1.drop_duplicates()
# 
df1 = df1.dropna(axis=1, how='all')
# # 10
df1 = df1[:15]

#X,Y
x=[x for x in df1['']][::-1]
y=[float(y) for y in df1['']][::-1]

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
plt.subplots(1, 1, figsize=(10, 6), dpi=100)

# 
plt.plot(x, y, color='red', marker='o', linestyle='-', alpha=0.5)

# 
plt.xlabel('', fontsize='16')
plt.ylabel('', fontsize='16')

# 
for x_, y_ in zip(x, y):
    plt.text(x_, y_, y_, ha='left', va='bottom')
```
